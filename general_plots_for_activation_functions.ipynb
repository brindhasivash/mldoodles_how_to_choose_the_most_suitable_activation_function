## General Plots for Activation Functions
#### Step, Sigmoid, tanh, ReLU, Leaky ReLU
These plots are used in https://mldoodles.com/how-to-choose-the-most-suitable-activation-function/

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(x):
    out = []
    for i in x:
        if i<0:
            out.append(0)
        else:
            out.append(1)
    return out
```

```python
inp = np.arange(-6, 6, 0.01)
step_out = step(inp)
```

```python
plt.plot(inp, step_out)
plt.title("Step Function")
plt.xlabel("input")
plt.ylabel("output")
#plt.grid()
plt.show()
```

```python
def sigmoid(x):
    out = []
    for i in x:
        out.append(1/(1+np.exp(-i)))
    return out
```

```python
inp = np.arange(-10, 10, 0.01)
sig_out = sigmoid(inp)
```

```python
plt.plot(inp, sig_out)
plt.title("Sigmoid Function")
plt.xlabel("input")
plt.ylabel("output")
#plt.grid()
plt.show()
```

```python
def tanh(x):
    x = np.array(x)
    tanh = (np.exp(x)-np.exp(-x)) / (np.exp(x)+np.exp(-x))
    # The below formula for tanh is identical to the above one. So use either one formula.
    # tanh = (2 / (1+np.exp(-2*x))) - 1 
    return list(tanh)
```

```python
inp = np.arange(-10, 10, 0.01)
tan_out = tanh(inp)
```

```python
plt.plot(inp, tan_out)
plt.title("tanh function")
plt.xlabel("input")
plt.ylabel("output")
#plt.grid()
plt.show()
```

```python
def relu(x):
    out = []
    for i in x:
        out.append(max(0, i))
    return out
```

```python
inp = np.arange(-10, 10, 0.01)
relu_out = relu(inp)
```

```python
plt.plot(inp, relu_out)
plt.title("ReLU Function")
plt.xlabel("input")
plt.ylabel("output")
#plt.grid()
plt.show()
```

```python
def l_relu(x): # leaky relu
    out = []
    for i in x:
        out.append(max(0.05*i, i))
    return out
```

```python
inp = np.arange(-10, 10, 0.01)
l_relu_out = l_relu(inp)
```

```python
plt.plot(inp, l_relu_out)
plt.title("Leaky ReLU Function")
plt.xlabel("input")
plt.ylabel("output")
#plt.grid()
plt.show()
```
