## Why tanh has stronger gradients compared to sigmoid?
Refer to https://mldoodles.com/how-to-choose-the-most-suitable-activation-function/ for explanation to below code.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    out = []
    grad = []
    for i in x:
        sig = 1/(1+np.exp(-i))
        g = sig * (1-sig) ## gradient
        out.append(sig)
        grad.append(g)
    return out, grad #returns output and gradient

def tanh(x):
    out = []
    grad = []
    for i in x:
        tan = (np.exp(i)-np.exp(-i)) / (np.exp(i)+np.exp(-i))
        g = 1 - (tan * tan) ## gradient
        out.append(tan)
        grad.append(g)
    return out, grad #returns output and gradient
```

```python
inp = np.arange(-6, 6, 0.01)

s_out1, sgrad1 = sigmoid(inp)
s_out2, sgrad2 = sigmoid(s_out1)
s_out3, sgrad3 = sigmoid(s_out2)

t_out1, tgrad1 = tanh(inp)
t_out2, tgrad2 = tanh(t_out1)
t_out3, tgrad3 = tanh(t_out2)
```

```python
# Plot for sigmoid

plt.figure(figsize=(16,5))

plt.subplot(131)
plt.axis([-7, 7, -0.2, 1.2])
plt.plot(inp, s_out1)
plt.plot(inp, sgrad1)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer1")
plt.legend(["sigmoid", "gradient"])

plt.subplot(132)
plt.axis([-7, 7, -0.2, 1.2])
plt.plot(inp, s_out2)
plt.plot(inp, sgrad2)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer2")
plt.legend(["sigmoid", "gradient"])

plt.subplot(133)
plt.axis([-7, 7, -0.2, 1.2])
plt.plot(inp, s_out3)
plt.plot(inp, sgrad3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer3")
plt.legend(["sigmoid", "gradient"])

plt.suptitle("Sigmoid and its gradient over 3 layers")
```

```python
# Plot for tanh
plt.figure(figsize=(16,5))

plt.subplot(131)
plt.axis([-7, 7, -1.2, 1.2])
plt.plot(inp, t_out1)
plt.plot(inp, tgrad1)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer1")
plt.legend(["tanh", "gradient"])

plt.subplot(132)
plt.axis([-7, 7, -1.2, 1.2])
plt.plot(inp, t_out2)
plt.plot(inp, tgrad2)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer2")
plt.legend(["tanh", "gradient"])

plt.subplot(133)
plt.axis([-7, 7, -1.2, 1.2])
plt.plot(inp, t_out3)
plt.plot(inp, tgrad3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Layer3")
plt.legend(["tanh", "gradient"])

plt.suptitle("tanh and its gradient over 3 layers")
```

```python
sigmoid_grad_range = [max(sgrad1)-min(sgrad1), max(sgrad2)-min(sgrad2), max(sgrad3)-min(sgrad3)]
tanh_grad_range = [max(tgrad1)-min(tgrad1), max(tgrad2)-min(tgrad2), max(tgrad3)-min(tgrad3)]
```

```python
# Plot sigmoid vs tanh gradients

ind = np.arange(3)
width = 0.35
plt.bar(ind, sigmoid_grad_range, width, label="sigmoid gradient range")
plt.bar(ind+width, tanh_grad_range, width, label = "tanh gradient range")
plt.ylabel("range")
plt.title("Gradient range comparison - sigmoid vs tanh")

plt.xticks(ind + width / 2, ('Layer1', 'Layer2', 'Layer3'))
plt.legend(loc='best')
plt.show()
```
